## Extract SIDs

```python
df = pd.read_csv(r"C:\Users\Admin\Documents\Studio Code\customer_retention_data\zoom_logs.csv", dtype={'SID': str})
df
```

```python
#Extract 5-digit SID from log in zoom name in case the student does not log in with verified school email
mask = (df['role'] == 'Student') & (df['SID_str'].isna())
extracted = (
    df.loc[mask, 'Name (original name)']
      .astype(str)
      .str.extract(r'(?<!\d)(\d{5}|\d{8})(?!\d)', expand=False)
)

# Assign only where we actually found a match
df.loc[mask, 'SID_str'] = extracted

#Check the newly extracted rows
df1 = df.loc[mask]
df1.head()
```

```python
# Identify rows where SID_str is exactly 8 digits (no letters or symbols)
mask_8 = df['SID_str'].astype(str).str.match(r'^\d{8}$', na=False)

# Transfer only those 8-digit numeric strings to SID
df.loc[mask_8, 'SID'] = df.loc[mask_8, 'SID_str']

# (Optional) Clear them from SID_str after moving
# df.loc[mask_8, 'SID_str'] = pd.NA
```

```python
df['SID_5digit'] = df['SID_str'].astype(str).str.extract(r'(\d{5})$')
df.head()
```

## Match Student List

```python
#read the list of students dataset
df_students = pd.read_excel(r"C:\Users\Admin\Documents\Studio Code\customer_retention_data\goitenemmai\SY25_MVA_Goitenemmai (1).xlsx", dtype={'SID_5digit': str, 'SID_8digit': str, 'SID_5digit': str})
df_students.head()
```

```python
m_full = df_merged.merge(df_students, left_on='SID', right_on='SID_8digit', how='left', indicator='match8')
m_full.head()
```

```python
m_full.columns
```

```python
unmatched = m_full[m_full['match8']!='both']
m5 = unmatched.merge(df_students, left_on=['SID_5digit_x','Grade_x'], right_on=['SID_5digit','Grade'], how='left', indicator='match5')
m5
```

```python
df_merged['role'].value_counts()
```

```python
m8 = df_merged.merge(
    df_students,
    left_on='SID',
    right_on='SID_8digit',
    how='left',
    indicator='merge8'
)
m8['match_level'] = m8['merge8'].map({'both': '8digit', 'left_only': pd.NA})
m8.drop(columns='merge8', inplace=True)

# ---------- 4️⃣ Pass 2: 5-digit + Grade for unmatched ----------
unmatched8 = m8[m8['match_level'].isna()].copy()
m5 = unmatched8.merge(
    df_students,
    left_on=['SID_5digit_x', 'Grade_x'],
    right_on=['SID_5digit', 'Grade'],
    how='left',
    indicator='merge5'
)
m5['match_level'] = m5['merge5'].map({'both': '5digit+grade', 'left_only': pd.NA})
m5.drop(columns='merge5', inplace=True)

# Combine back 8-digit matches + 5-digit matches
matched_partial = pd.concat([m8[m8['match_level']=='8digit'], m5], ignore_index=True)
```

```python
from fuzzywuzzy import process
```

```python
unmatched5 = matched_partial[matched_partial['match_level'].isna()].copy()

# build lookup dictionary of student names by grade
student_names = df_students.groupby('Grade')['Name'].apply(list).to_dict()

def fuzzy_match(row):
    name = str(row.get('Name (original name)', '')).strip()
    grade = str(row.get('Grade', '')).strip()
    if grade not in student_names or not name:
        return pd.NA
    best, score = process.extractOne(name, student_names[grade])
    return 'name_fuzzy' if score >= 90 else pd.NA

unmatched5['match_level'] = unmatched5.apply(fuzzy_match, axis=1)
```

```python
final_df = pd.concat([matched_partial[matched_partial['match_level'].notna()], unmatched5], ignore_index=True)
final_df['match_level'] = final_df['match_level'].fillna('none')

print(final_df['match_level'].value_counts())
```

```python
final_df.columns
```

```python
coverage = (
    final_df.groupby(['SID', 'match_level'])
             .size()
             .reset_index(name='records')
)
coverage_summary = coverage.groupby('match_level')['SID'].nunique().reset_index(name='unique_students')

print(coverage_summary)
```
